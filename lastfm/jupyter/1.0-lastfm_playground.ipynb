```python
import pylast
from datetime import datetime
import json
from os.path import join
from glob import glob
import pandas as pd

from lastfm import ROOT
from lastfm.src.authentication import authenticate
```

```python
RAW_DATA = join(ROOT, 'data', 'raw')
```

```python
files = glob(f'{RAW_DATA}/*.json', recursive=True)
file = files[0]
```

```python
to_concat = []
for file in files:
    with open(file, 'r') as f:
        data = json.load(f)
        data = json.loads(data)
        to_concat.append(pd.DataFrame(data))
df = pd.concat(to_concat)
```

```python
df = df.reset_index(drop=True)
df.timestamp = df.timestamp.astype(int)
df = df.sort_values('timestamp')
df.head(100)
df.query('timestamp > 1293836400') # 1293836400 = datetime(2011, 1, 1) the year scrobbling started
```
